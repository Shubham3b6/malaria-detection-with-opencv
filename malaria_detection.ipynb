### Calculate the area in the images

```python
file_path='C:\\Users\\shubham\\EXCEL OR CSV FILE\\cell_images\\Uninfected'
label = "Parasitized"

dirlist=glob.glob(os.path.join(file_path,'*.png'))
label='Uninfected'
file=[]
for img_path in dirlist:

    im = cv2.imread(img_path)

    im = cv2.GaussianBlur(im,(5,5),2)



    im_gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

    ret,thresh = cv2.threshold(im_gray,127,255,0)
    contours,_ = cv2.findContours(thresh,1,2)

    for contour in contours:
        cv2.drawContours(im_gray, contours, -1, (0,255,0), 3)


    li=[]
    li.append(label)     # Uninfected
    for i in range(5):
        try:
            area = cv2.contourArea(contours[i])

            
        except:
             
            li.append(0)
            
        else:
            li.append(area)
            
    
    file.append(li)
    print(li)

    
```

```python
file_path='C:\\Users\\shubham\\EXCEL OR CSV FILE\\cell_images\\Parasitized'
label2 = "Parasitized"

dirlist=glob.glob(os.path.join(file_path,'*.png'))

Area=[]
for img_path in dirlist:

    im = cv2.imread(img_path)

    im = cv2.GaussianBlur(im,(5,5),2)



    im_gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

    ret,thresh = cv2.threshold(im_gray,127,255,0)
    contours,_ = cv2.findContours(thresh,1,2)

    for contour in contours:
        cv2.drawContours(im_gray, contours, -1, (0,255,0), 3)


    li=[]
    li.append(label2) #Infected
    for i in range(5):
        try:
            area = cv2.contourArea(contours[i])

            
        except:
             
            li.append(0)
            
        else:
            li.append(area)
            
    
    Area.append(li)
    print(li)

    
```

```python
img.drop(['Unnamed: 0'],axis=1,inplace=True)
```

### Data Preprocessing

```python
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()

for i in img.columns:
    if isinstance(img[i][0], str):
            img[i] = encoder.fit_transform(img[i])
```

```python
X = img.drop(['target'], axis=1)
y = img['target']
```

### split the data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.2, random_state = 1)
```

### Model_1

```python
from sklearn import tree
#Assumed you have, X (predictor) and Y (target) for training data set and x_test(predictor) of test_dataset
# Create tree object 
model = tree.DecisionTreeClassifier() # for classification, here you can change the algorithm as gini or entropy (information gain) by default it is gini  
# model = tree.DecisionTreeRegressor() for regression
# Train the model using the training sets and check score
model.fit(X, y)
model.score(X, y)
#Predict Output
predicted= model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
DecisionTree=accuracy_score(y_test,predicted)*100
DecisionTree
```

```python
confusion_matrix(y_test,predicted)
```
